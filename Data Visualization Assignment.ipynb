```python
# Q 4:- Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
categories = ['A', 'B', 'C']
data = {
    'Category': np.random.choice(categories, 300),
    'Value': np.random.normal(loc=50, scale=10, size=300)
}
df = pd.DataFrame(data)

sns.set(style='whitegrid')
sns.boxplot(data=df, x='Category', y='Value', palette='Set2')

plt.title('Distribution of Values Across Categories')
plt.xlabel('Category')
plt.ylabel('Value')

plt.show()
```

```python
# Q 5 :- Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)

size = 300
x1 = np.random.normal(0, 1, size)
x2 = x1 * 0.8 + np.random.normal(0, 0.3, size)
x3 = x1 * -0.5 + np.random.normal(0, 0.5, size)
x4 = np.random.normal(0, 1, size)

df = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})

corr_matrix = df.corr()

sns.set(style='white')
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5, square=True)

plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Q 3:-  Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

sales_pivot = df.pivot_table(index='Month', columns='Day', values='Sales', aggfunc='mean')

month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales_pivot = sales_pivot.reindex(month_order)

plt.figure(figsize=(14, 6))
sns.heatmap(sales_pivot, cmap='YlGnBu', annot=False, linewidths=0.5, linecolor='gray')

plt.title('Heatmap of Sales by Month and Day')
plt.xlabel('Day of the Month')
plt.ylabel('Month')

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)
ax.set_title('3D Surface Plot of $z = \sin(\sqrt{x^2 + y^2})$')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    df['GDP'],
    df['Population'],
    s=df['Population'] * 2,
    alpha=0.6,
    c='skyblue',
    edgecolors='black',
    linewidth=1.5
)

for i, row in df.iterrows():
    plt.text(row['GDP'], row['Population'] + 20, row['Country'],
             ha='center', va='bottom', fontsize=9)

plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Bubble Chart: GDP vs Population by Country')

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Q 1 :- Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

x = np.linspace(0, 10, 500)
y = np.sin(x)


p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)',
           plot_width=700, plot_height=400)

p.line(x, y, legend_label="sin(x)", line_width=2, line_color="navy")


show(p)
```

````python
# Q 1 :-  Generate a list of 100 integers containing values between 90 to 130 and store it in the variable `int_list`. 
# After generating the list, find the following:
# (i) Write a Python function to calculate the mean of a given list of numbers. Create a function to find the median of a list of numbers.
# (ii) Develop a program to compute the mode of a list of integers.
# (iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights.
# (iv) Write a Python function to find the geometric mean of a list of positive numbers.
# (v) Create a program to calculate the harmonic mean of a list of values.
# (vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).
# (vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of outliers.
````

````python
import random
import statistics
import math

# Generate list of 100 integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# (i) Function to calculate the mean and median
def calculate_mean(data):
    return sum(data) / len(data)

def calculate_median(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    mid = n // 2
    if n % 2 == 0:
        return (sorted_data[mid - 1] + sorted_data[mid]) / 2
    else:
        return sorted_data[mid]

# (ii) Function to calculate the mode
def calculate_mode(data):
    freq = {}
    for num in data:
        freq[num] = freq.get(num, 0) + 1
    max_freq = max(freq.values())
    mode = [k for k, v in freq.items() if v == max_freq]
    return mode

# (iii) Function to calculate the weighted mean
def calculate_weighted_mean(values, weights):
    weighted_sum = sum(v * w for v, w in zip(values, weights))
    total_weight = sum(weights)
    return weighted_sum / total_weight

# (iv) Function to calculate geometric mean
def calculate_geometric_mean(data):
    product = 1
    for num in data:
        product *= num
    return product ** (1 / len(data))

# (v) Function to calculate harmonic mean
def calculate_harmonic_mean(data):
    return len(data) / sum(1 / x for x in data)

# (vi) Function to calculate midrange
def calculate_midrange(data):
    return (min(data) + max(data)) / 2

# (vii) Function to calculate trimmed mean
def calculate_trimmed_mean(data, trim_percent):
    sorted_data = sorted(data)
    n = len(data)
    trim_count = int(n * trim_percent / 100)
    trimmed_data = sorted_data[trim_count:n - trim_count]
    return sum(trimmed_data) / len(trimmed_data)

# Example usage:
print("Mean:", calculate_mean(int_list))
print("Median:", calculate_median(int_list))
print("Mode:", calculate_mode(int_list))
# Weighted mean: use arbitrary weights for demonstration
weights = [random.randint(1, 5) for _ in range(100)]
print("Weighted Mean:", calculate_weighted_mean(int_list, weights))
print("Geometric Mean:", calculate_geometric_mean(int_list))
print("Harmonic Mean:", calculate_harmonic_mean(int_list))
print("Midrange:", calculate_midrange(int_list))
print("Trimmed Mean (10%):", calculate_trimmed_mean(int_list, 10))
````

````python
# 2 :-  Generate a list of 500 integers containing values between 200 to 300 and store it in the variable `int_list2`.
# After generating the list, find the following:
# (i) Compare the given list of visualization for the given data:
# 1. Frequency & Gaussian distribution
# 2. Frequency smoothened KDE plot
# 3. Gaussian distribution & smoothened KDE plot
# (ii) Write a Python function to calculate the range of a given list of numbers.
# (iii) Create a program to find the variance and standard deviation of a list of numbers.
# (iv) Implement a function to compute the interquartile range (IQR) of a list of values.
# (v) Build a program to calculate the coefficient of variation for a dataset.
# (vi) Write a Python function to find the mean absolute deviation (MAD) of a list of numbers.
# (vii) Create a program to calculate the quartile deviation of a list of values.
# (viii) Implement a function to find the range-based coefficient of dispersion for a dataset.
````

````python
#Ans :- 
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate list of 500 integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# --- (i) Visualizations ---

def plot_visualizations(data):
    plt.figure(figsize=(18, 5))

    # 1. Frequency & Gaussian Distribution
    plt.subplot(1, 3, 1)
    sns.histplot(data, kde=False, bins=20, color='skyblue', stat='density', label='Histogram')
    mean = np.mean(data)
    std = np.std(data)
    x = np.linspace(min(data), max(data), 100)
    y = (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std) ** 2)
    plt.plot(x, y, label='Gaussian Distribution', color='red')
    plt.title("Frequency & Gaussian Distribution")
    plt.legend()

    # 2. Frequency smoothened KDE plot
    plt.subplot(1, 3, 2)
    sns.kdeplot(data, shade=True, color='green', label='KDE')
    plt.title("Frequency Smoothened KDE Plot")
    plt.legend()

    # 3. Gaussian & KDE Plot
    plt.subplot(1, 3, 3)
    sns.kdeplot(data, color='blue', label='KDE')
    plt.plot(x, y, label='Gaussian Distribution', color='orange')
    plt.title("Gaussian Distribution & KDE Plot")
    plt.legend()

    plt.tight_layout()
    plt.show()

plot_visualizations(int_list2)

# --- (ii) Range ---
def calculate_range(data):
    return max(data) - min(data)

# --- (iii) Variance & Standard Deviation ---
def calculate_variance_std(data):
    mean = sum(data) / len(data)
    squared_diffs = [(x - mean) ** 2 for x in data]
    variance = sum(squared_diffs) / len(data)
    std_dev = variance ** 0.5
    return variance, std_dev

# --- (iv) Interquartile Range (IQR) ---
def calculate_iqr(data):
    q75, q25 = np.percentile(data, [75 ,25])
    return q75 - q25

# --- (v) Coefficient of Variation ---
def coefficient_of_variation(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    return (std_dev / mean) * 100

# --- (vi) Mean Absolute Deviation (MAD) ---
def mean_absolute_deviation(data):
    mean = sum(data) / len(data)
    return sum(abs(x - mean) for x in data) / len(data)

# --- (vii) Quartile Deviation ---
def quartile_deviation(data):
    q1, q3 = np.percentile(data, [25, 75])
    return (q3 - q1) / 2

# --- (viii) Range-based Coefficient of Dispersion ---
def range_based_dispersion(data):
    return (max(data) - min(data)) / (max(data) + min(data))

# --- Output Results ---
print("Range:", calculate_range(int_list2))
variance, std_dev = calculate_variance_std(int_list2)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
print("Interquartile Range (IQR):", calculate_iqr(int_list2))
print("Coefficient of Variation (%):", coefficient_of_variation(int_list2))
print("Mean Absolute Deviation (MAD):", mean_absolute_deviation(int_list2))
print("Quartile Deviation:", quartile_deviation(int_list2))
print("Range-based Coefficient of Dispersion:", range_based_dispersion(int_list2))
````

````python
# 3 :- )  Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.
````

````python
#Ans :-
class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        if len(values) != len(probabilities):
            raise ValueError("Values and probabilities must be of the same length.")
        if not abs(sum(probabilities) - 1.0) < 1e-6:
            raise ValueError("Probabilities must sum to 1.")

        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        return sum(v * p for v, p in zip(self.values, self.probabilities))

    def variance(self):
        mean = self.expected_value()
        return sum(((v - mean) ** 2) * p for v, p in zip(self.values, self.probabilities))

    def __repr__(self):
        return f"DiscreteRandomVariable(values={self.values}, probabilities={self.probabilities})"
````

````python
# Define a discrete random variable X with outcomes and their probabilities
values = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.25, 0.15]

# Create the object
X = DiscreteRandomVariable(values, probabilities)

# Compute expected value and variance
print("Expected Value (E[X]):", X.expected_value())
print("Variance (Var[X]):", X.variance())
````

````python
# 4 :- Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and variance of the outcomes.
````

````python
#Ans :-
import random

def simulate_die_rolls(n_rolls=100000):
    outcomes = [random.randint(1, 6) for _ in range(n_rolls)]
    return outcomes

def expected_value(data):
    return sum(data) / len(data)

def variance(data):
    mean = expected_value(data)
    return sum((x - mean) ** 2 for x in data) / len(data)

# Simulate 100,000 rolls of a fair 6-sided die
rolls = simulate_die_rolls()

# Calculate expected value and variance
mean = expected_value(rolls)
var = variance(rolls)

print(f"Simulated Expected Value: {mean:.4f}")
print(f"Simulated Variance: {var:.4f}")
````

````python
# 5 :- Create a Python function to generate random samples from a given probability distribution 
#(e.g., binomial, Poisson) and calculate their mean and variance.
````

````python
# Ans :- 
import numpy as np

def sample_and_analyze(distribution, size=1000, **kwargs):
    if distribution == 'binomial':
        # Required kwargs: n (trials), p (probability)
        samples = np.random.binomial(n=kwargs['n'], p=kwargs['p'], size=size)
    elif distribution == 'poisson':
        # Required kwarg: lam (lambda)
        samples = np.random.poisson(lam=kwargs['lam'], size=size)
    else:
        raise ValueError("Unsupported distribution. Use 'binomial' or 'poisson'.")

    mean = np.mean(samples)
    variance = np.var(samples)
    return samples, mean, variance
````

````python
samples, mean, var = sample_and_analyze('binomial', size=10000, n=10, p=0.5)
print("Binomial Sample Mean:", mean)
print("Binomial Sample Variance:", var)
````

````python
samples, mean, var = sample_and_analyze('poisson', size=10000, lam=4)
print("Poisson Sample Mean:", mean)
print("Poisson Sample Variance:", var)
````

````python
# 6 :- Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute 
#the mean, variance, and standard deviation of the samples.
````

````python
#Ans :- 
import numpy as np

def generate_normal_samples(mean=0, std_dev=1, size=1000):
    """
    Generates random numbers from a Gaussian distribution.
    
    Parameters:
        mean (float): Mean of the distribution
        std_dev (float): Standard deviation of the distribution
        size (int): Number of samples to generate
        
    Returns:
        samples (np.ndarray): Array of generated samples
        sample_mean (float): Mean of samples
        sample_variance (float): Variance of samples
        sample_std_dev (float): Standard deviation of samples
    """
    samples = np.random.normal(loc=mean, scale=std_dev, size=size)
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples)
    sample_std_dev = np.std(samples)
    
    return samples, sample_mean, sample_variance, sample_std_dev

# Example: Generate samples from N(5, 2^2)
samples, mean, var, std = generate_normal_samples(mean=5, std_dev=2, size=10000)

print(f"Sample Mean: {mean:.4f}")
print(f"Sample Variance: {var:.4f}")
print(f"Sample Standard Deviation: {std:.4f}")
````

````python
# Q 7:- Use seaborn library to load tips dataset. Find the following from the dataset for the columns total_bill
# and tip`:
# (i) Write a Python function that calculates their skewness.
# (ii) Create a program that determines whether the columns exhibit positive skewness, negative skewness, or is
# approximately symmetric.
# (iii) Write a function that calculates the covariance between two columns.
# (iv) Implement a Python program that calculates the Pearson correlation coefficient between two columns.
# (v) Write a script to visualize the correlation between two specific columns in a Pandas DataFrame using
# scatter plots.
````

````python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# Load the tips dataset
df = sns.load_dataset('tips')

# --- (i) Function to calculate skewness ---
def calculate_skewness(series):
    return skew(series)

# --- (ii) Determine skewness type ---
def skewness_type(series):
    sk = calculate_skewness(series)
    if sk > 0:
        return "Positive Skewness"
    elif sk < 0:
        return "Negative Skewness"
    else:
        return "Approximately Symmetric"

# --- (iii) Calculate covariance ---
def calculate_covariance(series1, series2):
    return np.cov(series1, series2, ddof=0)[0][1]  # population covariance

# --- (iv) Calculate Pearson correlation ---
def calculate_pearson_corr(series1, series2):
    return np.corrcoef(series1, series2)[0][1]

# --- (v) Scatter plot for visualization ---
def plot_scatter(x, y, x_label, y_label):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=x, y=y)
    plt.title(f'Scatter Plot: {x_label} vs {y_label}')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True)
    plt.show()

# Columns
total_bill = df['total_bill']
tip = df['tip']

# Results
print("Skewness of total_bill:", calculate_skewness(total_bill))
print("Skewness of tip:", calculate_skewness(tip))
print("Skewness type of total_bill:", skewness_type(total_bill))
print("Skewness type of tip:", skewness_type(tip))
print("Covariance between total_bill and tip:", calculate_covariance(total_bill, tip))
print("Pearson Correlation Coefficient between total_bill and tip:", calculate_pearson_corr(total_bill, tip))

# Plot
plot_scatter(df['total_bill'], df['tip'], 'Total Bill', 'Tip')
````

````python
#Q 8:- Write a Python function to calculate the probability density function (PDF) of a continuous random
#variable for a given normal distribution. 
````

````python
import math

def normal_pdf(x, mean=0, std_dev=1):
   
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    exponent = math.exp(-((x - mean) ** 2) / (2 * std_dev ** 2))
    return coefficient * exponent
````

````python
# Example: PDF of N(0, 1) at x = 1
pdf_value = normal_pdf(1, mean=0, std_dev=1)
print(f"PDF at x=1 for N(0,1): {pdf_value:.6f}")
````

````python
#Q 9 :- Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.
````

````python
import math

def exponential_cdf(x, rate):
    
    if x < 0:
        return 0.0  # CDF is 0 for x < 0
    if rate <= 0:
        raise ValueError("Rate parameter λ must be positive.")
    
    return 1 - math.exp(-rate * x)
````

````python
x_value = 2.0
lambda_rate = 1.5

cdf_result = exponential_cdf(x_value, lambda_rate)
print(f"CDF at x = {x_value} with λ = {lambda_rate}: {cdf_result:.4f}")
````

````python
#Q 10 :- Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.
````

````python
import math

def poisson_pmf(k, lam):
  
    if k < 0 or not isinstance(k, int):
        raise ValueError("k must be a non-negative integer.")
    if lam <= 0:
        raise ValueError("Lambda (λ) must be positive.")

    return (math.exp(-lam) * lam**k) / math.factorial(k)
````

````python
k = 3       # Number of occurrences
lam = 2.5   # Average rate (λ)

pmf_val = poisson_pmf(k, lam)
print(f"P(X = {k}) with λ = {lam} is {pmf_val:.4f}")
````

````python
#Q 11 :-  A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors
# who make a purchase). They collect data from the old and new layouts to compare.
# To generate the data use the following command:
# ```python
# import numpy as np
# 50 purchases out of 1000 visitors
# old_layout = np.array([1] * 50 + [0] * 950)
# 70 purchases out of 1000 visitors  
# new_layout = np.array([1] * 70 + [0] * 930)
# Apply z-test to find which layout is successful.
````

````python
import numpy as np

old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

p1 = old_layout.mean()  # conversion rate old layout
p2 = new_layout.mean()  # conversion rate new layout

n1 = len(old_layout)
n2 = len(new_layout)

print(f"Old conversion rate: {p1:.4f}")
print(f"New conversion rate: {p2:.4f}")
````

````python
import numpy as np
from scipy.stats import norm

old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

p1 = old_layout.mean()
p2 = new_layout.mean()

n1 = len(old_layout)
n2 = len(new_layout)

# Pooled proportion
p_pool = (old_layout.sum() + new_layout.sum()) / (n1 + n2)

# Standard error
se = np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))

# z-score
z = (p2 - p1) / se

# p-value for one-tailed test (new_layout better than old_layout)
p_value = 1 - norm.cdf(z)

print(f"Old layout conversion rate: {p1:.4f}")
print(f"New layout conversion rate: {p2:.4f}")
print(f"Z-score: {z:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("The new layout leads to a significantly higher conversion rate (reject H0).")
else:
    print("No significant difference in conversion rates (fail to reject H0).")
````

````python
#Q 15 :- A company is testing two different website layouts to see which one leads to higher click-through rates.
# Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of
# freedom, and p-value.
# Use the following data:
# ```python
# layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
# layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]
````

````python
import numpy as np
from scipy.stats import t

def ab_test(layout_a, layout_b):
    n1, n2 = len(layout_a), len(layout_b)
    mean1, mean2 = np.mean(layout_a), np.mean(layout_b)
    var1, var2 = np.var(layout_a, ddof=1), np.var(layout_b, ddof=1)

    # t-statistic
    se = np.sqrt(var1/n1 + var2/n2)
    t_stat = (mean1 - mean2) / se

    # degrees of freedom (Welch-Satterthwaite equation)
    df = (var1/n1 + var2/n2)**2 / ((var1**2)/((n1**2)*(n1-1)) + (var2**2)/((n2**2)*(n2-1)))

    # two-tailed p-value
    p_value = 2 * t.sf(np.abs(t_stat), df)

    print(f"Mean (Layout A): {mean1:.2f}")
    print(f"Mean (Layout B): {mean2:.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"Degrees of freedom: {df:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < 0.05:
        print("There is a significant difference in click-through rates (reject H0).")
    else:
        print("No significant difference in click-through rates (fail to reject H0).")

# Sample data
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Run test
ab_test(layout_a_clicks, layout_b_clicks)
````

````python
import numpy as np
from scipy.stats import t

def cholesterol_treatment_test(existing, new):
    n1, n2 = len(existing), len(new)
    mean1, mean2 = np.mean(existing), np.mean(new)
    var1, var2 = np.var(existing, ddof=1), np.var(new, ddof=1)

    # t-statistic
    se = np.sqrt(var1/n1 + var2/n2)
    t_stat = (mean1 - mean2) / se

    # degrees of freedom (Welch's approximation)
    df = (var1/n1 + var2/n2)**2 / ((var1**2)/((n1**2)*(n1-1)) + (var2**2)/((n2**2)*(n2-1)))

    # One-tailed p-value (H1: new drug < existing drug, so mean2 < mean1)
    p_value = t.sf(t_stat, df)  # sf because t_stat is positive (mean1 > mean2)

    print(f"Mean (Existing drug): {mean1:.2f}")
    print(f"Mean (New drug): {mean2:.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"Degrees of freedom: {df:.4f}")
    print(f"P-value (one-tailed): {p_value:.4f}")

    if p_value < 0.05:
        print("New drug is significantly more effective in reducing cholesterol (reject H0).")
    else:
        print("No significant evidence that the new drug is more effective (fail to reject H0).")

# Clinical trial data
existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

# Run test
cholesterol_treatment_test(existing_drug_levels, new_drug_levels)
````

````python
# Q 17:- A school district introduces an educational intervention program to improve math scores. Write a Python
# function to analyze pre- and post-intervention test scores, calculating the t-statistic and p-value to
# determine if the intervention had a significant impact.
# Use the following data of test score:
# ```python
# pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
# post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
# ```
````

````python
import numpy as np
from scipy.stats import t

def test_intervention_effect(pre_scores, post_scores):
    # Calculate differences
    differences = np.array(post_scores) - np.array(pre_scores)
    n = len(differences)
    mean_diff = np.mean(differences)
    std_diff = np.std(differences, ddof=1)

    # t-statistic
    t_stat = mean_diff / (std_diff / np.sqrt(n))

    # Degrees of freedom
    df = n - 1

    # One-tailed p-value
    p_value = 1 - t.cdf(t_stat, df)

    print(f"Mean Difference: {mean_diff:.4f}")
    print(f"Standard Deviation of Differences: {std_diff:.4f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"Degrees of Freedom: {df}")
    print(f"P-value (one-tailed): {p_value:.4f}")

    if p_value < 0.05:
        print("There is significant evidence the intervention improved math scores (reject H0).")
    else:
        print("No significant evidence that the intervention improved math scores (fail to reject H0).")

# Sample data
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Run the analysis
test_intervention_effect(pre_intervention_scores, post_intervention_scores)
````

````python
#Q 18 :- An HR department wants to investigate if there's a gender-based salary gap within the company. Develop
# a program to analyze salary data, calculate the t-statistic, and determine if there's a statistically
# significant difference between the average salaries of male and female employees.
# Use the below code to generate synthetic data:
# ```python
# Generate synthetic salary data for male and female employees
# np.random.seed(0)  # For reproducibility
# male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
# female_salaries = np.random.normal(loc=55000, scale=9000, size=20)
# ```
````

````python
import numpy as np
from scipy.stats import t

def salary_gap_analysis(male_salaries, female_salaries):
    n1, n2 = len(male_salaries), len(female_salaries)
    mean1, mean2 = np.mean(male_salaries), np.mean(female_salaries)
    var1, var2 = np.var(male_salaries, ddof=1), np.var(female_salaries, ddof=1)

    # t-statistic using Welch's t-test
    se = np.sqrt(var1/n1 + var2/n2)
    t_stat = (mean1 - mean2) / se

    # Degrees of freedom (Welch's approximation)
    df = (var1/n1 + var2/n2)**2 / ((var1**2)/((n1**2)*(n1-1)) + (var2**2)/((n2**2)*(n2-1)))

    # Two-tailed p-value
    p_value = 2 * t.sf(np.abs(t_stat), df)

    print(f"Mean Salary (Male): ${mean1:.2f}")
    print(f"Mean Salary (Female): ${mean2:.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"Degrees of Freedom: {df:.4f}")
    print(f"P-value (two-tailed): {p_value:.4f}")

    if p_value < 0.05:
        print("There is a statistically significant gender-based salary gap (reject H0).")
    else:
        print("No statistically significant salary gap detected (fail to reject H0).")

# Generate synthetic salary data
np.random.seed(0)
male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

# Run the analysis
salary_gap_analysis(male_salaries, female_salaries)
````

````python
# Q 19:- ? A manufacturer produces two different versions of a product and wants to compare their quality scores.
# Create a Python function to analyze quality assessment data, calculate the t-statistic, and decide
# whether there's a significant difference in quality between the two versions.
# Use the following data:
# ```python
# version1_scores = [85, 88, 82, 89, 87, 84, 90, 88, 85, 86, 91, 83, 87, 84, 89, 86, 84, 88, 85, 86, 89, 90, 87, 88, 85]
# version2_scores = [80, 78, 83, 81, 79, 82, 76, 80, 78, 81, 77, 82, 80, 79, 82, 79, 80, 81, 79, 82, 79, 78, 80, 81, 82]
# ```
````

````python
import numpy as np
from scipy.stats import t

def compare_quality(version1, version2):
    n1, n2 = len(version1), len(version2)
    mean1, mean2 = np.mean(version1), np.mean(version2)
    var1, var2 = np.var(version1, ddof=1), np.var(version2, ddof=1)

    # Calculate standard error
    se = np.sqrt(var1 / n1 + var2 / n2)

    # Compute t-statistic
    t_stat = (mean1 - mean2) / se

    # Degrees of freedom (Welch’s approximation)
    df = (var1/n1 + var2/n2)**2 / (
        ((var1**2) / ((n1**2) * (n1 - 1))) + ((var2**2) / ((n2**2) * (n2 - 1)))
    )

    # Two-tailed p-value
    p_value = 2 * t.sf(np.abs(t_stat), df)

    print(f"Mean Quality Score (Version 1): {mean1:.2f}")
    print(f"Mean Quality Score (Version 2): {mean2:.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"Degrees of Freedom: {df:.2f}")
    print(f"P-value (two-tailed): {p_value:.4f}")

    if p_value < 0.05:
        print("There is a significant difference in quality between the two versions (reject H0).")
    else:
        print("No significant difference in quality between the two versions (fail to reject H0).")

# Data
version1_scores = [85, 88, 82, 89, 87, 84, 90, 88, 85, 86, 91, 83, 87, 84, 89, 86, 84, 88, 85, 86, 89, 90, 87, 88, 85]
version2_scores = [80, 78, 83, 81, 79, 82, 76, 80, 78, 81, 77, 82, 80, 79, 82, 79, 80, 81, 79, 82, 79, 78, 80, 81, 82]

# Run the test
compare_quality(version1_scores, version2_scores)
````

````python
# Q 21:-  A political analyst wants to determine if there is a significant association between age groups and voter
# preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify
# them into different age groups and candidate preferences. Perform a Chi-Square test to determine if
# there is a significant association between age groups and voter preferences.
# Use the below code to generate data:
# ```python
# np.random.seed(0)
# age_groups = np.random.choice([ 18 30 , 31 50 , 51+', 51+'], size=30)
# voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)
````

````python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Fixing malformed strings
np.random.seed(0)

# Correct age groups
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=30)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

# Create a DataFrame
data = pd.DataFrame({
    'Age Group': age_groups,
    'Voter Preference': voter_preferences
})
````

````python
# Create a contingency table
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Perform the Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Contingency Table:")
print(contingency_table)
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns))

print(f"\nChi-Square Statistic: {chi2:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p:.4f}")

# Interpretation
if p < 0.05:
    print("\nThere is a significant association between age group and voter preference (reject H0).")
else:
    print("\nNo significant association between age group and voter preference (fail to reject H0).")
````

````python
#Q 34 :- . A company produces three different versions of a product: Standard, Premium, and Deluxe. The
# company wants to determine if there is a significant difference in customer satisfaction scores among the
# three product versions. They conducted a survey and collected customer satisfaction scores for each
# version from a random sample of customers. Perform an ANOVA test to determine if there is a significant
# difference in customer satisfaction scores.
# Use the following data:
# ```python
# Sample data: Customer satisfaction scores for each product version
# standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
# premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
# deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]
# ``
````
